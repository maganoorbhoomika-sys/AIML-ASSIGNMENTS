```python
import matplotlib.pyplot as plt
import numpy as np



```

```python
 ##Bar chart data (Sales of products)
products = ['A', 'B', 'C', 'D', 'E']
sales = [120, 150, 90, 200, 170]

```

```python
# Pie chart data (Market share)
market_share = [25, 30, 15, 20, 10]

```

```python
# Histogram data (Random dataset e.g. student marks)
marks = np.random.randint(40, 100, 100)







```

```python
 ##BAR CHART

plt.figure()
plt.bar(products, sales)
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.grid(True)
plt.show()



```

```python
# PIE CHART

plt.figure()
plt.pie(market_share, labels=products, autopct='%1.1f%%')
plt.title("Market Share Distribution")
plt.show()


```

```python
# HISTOGRAM

plt.figure()
plt.hist(marks, bins=10)
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
