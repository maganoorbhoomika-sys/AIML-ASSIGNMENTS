```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
import io

print("🔄 Click 'Choose Files' below to upload an image from your computer:")
print("-" * 60)

# 1. Native Colab File Uploader
uploaded = files.upload()

# Check if a file was uploaded
if not uploaded:
    print("\n❌ No file uploaded. Please re-run the cell and select an image.")
else:
    # 2. Grab the file name and content bytes
    file_name = list(uploaded.keys())[0]
    image_bytes = uploaded[file_name]

    # 3. Open image using PIL and convert to RGB format
    image = Image.open(io.BytesIO(image_bytes)).convert("RGB")

    # 4. Convert image to a raw numerical NumPy array (Tensor)
    img_matrix = np.array(image)
    height, width, channels = img_matrix.shape

    # --- METADATA OUTPUT ---
    print("\n" + "="*50)
    print(" 📊 IMAGE TENSOR METADATA")
    print("="*50)
    print(f"File Name            : {file_name}")
    print(f"Full Array Shape     : {img_matrix.shape} -> (Height, Width, Channels)")
    print(f"Total Pixels         : {height * width:,} pixels")
    print(f"Color Space Depth    : {channels} Channels (Red, Green, Blue)")
    print(f"Data Matrix Primitive: {img_matrix.dtype} (Unsigned 8-bit integers)")

    # --- NUMERICAL MATRIX ANALYSIS ---
    print("\n" + "="*50)
    print(" 🔢 PIXEL LEVEL GRID (Top-Left 3x3 Coordinate Box)")
    print("="*50)
    print("Each bracket contains [Red, Green, Blue] values scaled from 0 to 255:\n")

    # Extract the top left 3x3 patch
    top_left_patch = img_matrix[0:3, 0:3, :]
    print(top_left_patch)

    # --- MACHINE LEARNING PREPARATION EXAMPLE ---
    print("\n" + "="*50)
    print(" 🤖 ML INPUT NORMALIZATION SAMPLE")
    print("="*50)
    print("Neural Networks prefer normalized float values [0.0 to 1.0] instead of [0 to 255]:\n")
    normalized_patch = top_left_patch / 255.0
    print(np.round(normalized_patch, 3))

    # --- VISUAL CONFIRMATION ---
    plt.figure(figsize=(6, 6))
    plt.imshow(img_matrix)
    plt.title(f"Processed Tensor: {img_matrix.shape}", fontsize=12)
    plt.axis('off')
    plt.show()
```
