```python
# 1. Generate requirements.txt for Streamlit Cloud
# (Using headless prevents libGL server errors on Streamlit Cloud)
with open("requirements.txt", "w") as f:
    f.write('''streamlit>=1.30.0
opencv-python-headless>=4.8.0
numpy>=1.24.0
Pillow>=10.0.0
''')

# 2. Generate the production-ready app.py
with open("app.py", "w") as f:
    f.write('''import streamlit as st
import cv2
import numpy as np
from PIL import Image

# Page Configuration
st.set_page_config(page_title="OpenCV Filter Lab", page_icon="📷", layout="wide")

st.title("📷 OpenCV Image Filter Lab")
st.markdown("Upload an image and instantly apply mathematical matrix filters. No AI training required—just pure computer vision math!")

# File Uploader
uploaded_file = st.file_uploader("Upload an image (JPG, PNG)", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    # 1. Read Image using PIL, then convert to OpenCV format (NumPy Array)
    raw_image = Image.open(uploaded_file).convert('RGB')
    img_array = np.array(raw_image)

    # Sidebar Controls for the Lab
    st.sidebar.header("🎛️ Filter Controls")
    filter_type = st.sidebar.radio(
        "Choose a Filter:",
        ("Grayscale", "Gaussian Blur", "Canny Edge Detection")
    )

    # Process Image based on selection
    processed_img = None

    if filter_type == "Grayscale":
        # Math: Averages the RGB channels into a single intensity channel
        processed_img = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)

    elif filter_type == "Gaussian Blur":
        # Math: Averages local pixel neighborhoods to soften sharp transitions
        blur_intensity = st.sidebar.slider("Blur Intensity (Kernel Size)", min_value=3, max_value=51, value=15, step=2)
        processed_img = cv2.GaussianBlur(img_array, (blur_intensity, blur_intensity), 0)

    elif filter_type == "Canny Edge Detection":
        # Math: Calculates intensity gradients to find harsh lines and borders
        threshold1 = st.sidebar.slider("Min Threshold", 0, 255, 100)
        threshold2 = st.sidebar.slider("Max Threshold", 0, 255, 200)
        processed_img = cv2.Canny(img_array, threshold1, threshold2)

    # 3. UI Display: Before and After
    st.markdown("---")
    col1, col2 = st.columns(2)

    with col1:
        st.subheader("🖼️ Before (Original)")
        st.image(raw_image, use_container_width=True)

    with col2:
        st.subheader(f"✨ After ({filter_type})")
        # Streamlit handles both 1-channel (gray/edges) and 3-channel (color/blur) automatically
        st.image(processed_img, use_container_width=True, clamp=True)
''')

```
