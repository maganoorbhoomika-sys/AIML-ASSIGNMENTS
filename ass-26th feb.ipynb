Data Doctor
Description : Clean a dataset by handling missing values, removing duplicates, standardizing text, and explain why cleaning matters.

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "alice", "Charlie", None],
    "Age": [20, 21, 20, None, 22],
    "Marks": [85, 90, 85, 70, None]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
# Fill missing Age with average
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill missing Marks with 0
df['Marks'] = df['Marks'].fillna(0)

# Drop rows where Name is missing
df = df.dropna(subset=['Name'])
```

```python
df = df.drop_duplicates()
```

```python
df['Name'] = df['Name'].str.lower().str.strip()
```

```python
print("Cleaned Data:\n", df)
```
